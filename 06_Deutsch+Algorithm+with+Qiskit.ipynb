# Deutsch Algorithm with Qiskit

In the Deutsch‚ÄìJozsa problem, we are given a black box quantum computer known as an oracle that implements some function:

ùëì:{0,1}ùëõ‚Üí{0,1}


The function takes n-bit binary values as input and produces either a 0 or a 1 as output for each such value. We are promised that the function is either constant (0 on all inputs or 1 on all inputs) or balanced (1 for exactly half of the input domain and 0 for the other half). The task then is to determine if 
ùëì is constant or balanced by using the oracle.

## Importing necessary libraries 

```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram
```

## Constant Oracles - Oracle with Identity Gates

```python
# Implement Deutsch Algorithm for a Constant Function
qc = QuantumCircuit(2,1,name="qc")

# Place the second qubit in state |1>
qc.x(1)
qc.barrier()

# Place each qubit in Superposition 
qc.h(0)
qc.h(1)
qc.barrier()

# Creating a constant oracle . This is an example how to create an Oracle, 
#encoding an unknown function (here: constant function). 
#Aim: prove that the Oracle is a contsant function
qc.id(0)
qc.id(1)
qc.barrier()

# Putting a Hadamard to the first qubit
qc.h(0)
qc.barrier()

# Measuring the first qubit
qc.measure(0,0)

qc.draw('mpl')
```

```python
#Run this circuit
# Use Aer's qasm_simulator
backend = Aer.get_backend('qasm_simulator')

# Execute the circuit on the qasm simulator
result = backend.run(qc, shots=1000).result()
counts = result.get_counts(qc)

# Returns counts
print("\nTotal counts are:",counts)

# Plot a histogram
plot_histogram(counts)
```

The Deutsch algo has determined that the oracle is constant
